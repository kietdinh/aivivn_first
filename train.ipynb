```python
import pandas as pd
import numpy as np
```

```python
import re
```

```python
regep = re.compile(r"[-()\"#/@;:<>{}`+=~|.!?,]_*&^%$")
renumber = re.compile(r"\d+")
```

```python
def transform(df):
    df['content'] = df['content'].apply(lambda x: str(x).lower())
    df['content'] = df['content'].apply(lambda x: renumber.sub('ddddd', x))
    df['content'] = df['content'].apply(lambda x: regep.sub('', x))
    df['content'] = df['content'].apply(lambda x: ' '.join([s for s in x.split(' ') if len(s) >= 3]))
    return df
```

```python
data_train = pd.read_csv('train.csv')
data_train = data_train.sample(frac=1).reset_index(drop=True)
```

```python
data_train.groupby('label').agg('count')
```

```python
data_train = transform(data_train)
```

```python
from collections import defaultdict
```

```python
def query(df, text, show=0):
    label = defaultdict(int)
    for _, row in df.iterrows():
        if text in row['content']:
            label[row['label']] += 1
            if row['label'] == show:
                print(row['content'])
    print(label.items())
```

```python
n = len(data_train)
data_valid = data_train[:int(n*0.3)]
data_train = data_train[len(data_valid):]
```

```python
data_test = pd.read_csv('test.csv')
```

```python
data_test = transform(data_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
vectorizer = TfidfVectorizer(max_df=0.9,
#                              max_features=1000,
                             min_df=15,
                             use_idf=True,
                             ngram_range=(1,2))
```

```python
X_train = vectorizer.fit_transform(data_train['content'])
X_valid = vectorizer.transform(data_valid['content'])
X_test = vectorizer.transform(data_test['content'])
```

```python
X_train.shape
```

```python
vectorizer.get_feature_names()
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
import lightgbm
```

```python
clfs = {
    "RF": RandomForestClassifier(n_estimators=200, min_samples_leaf=2, max_depth=30, max_features=0.7, random_state=42, n_jobs=-1),
    "GB": GradientBoostingClassifier(n_estimators=200, min_samples_split=2, max_depth=10, max_features=0.7, random_state=111),
    "BB": BalancedBaggingClassifier(n_estimators=200, max_features=0.7, max_samples=0.7, random_state=111),
    "LR": LogisticRegression(penalty='l2', C=1.0),
}
```

```python
for name, clf in clfs.items():
    y_true = data_train['label']
    clf.fit(X_train, y_true)
    y_pred = clf.predict(X_train)
    
    print("F1", f1_score(y_true, y_pred))
    print("CM", confusion_matrix(y_true, y_pred))
    
    y_true = data_valid['label']
    y_pred = clf.predict(X_valid)
    
    print("F1", f1_score(y_true, y_pred))
    print("CM", confusion_matrix(y_true, y_pred))
```

```python
parameters = {
    'application': 'binary',
    'objective': 'binary',
    'metric': 'auc',
    'is_unbalance': 'true',
    'boosting': 'gbdt',
    'num_leaves': 31,
    'feature_fraction': 0.7,
    'bagging_fraction': 0.5,
    'bagging_freq': 30,
    'learning_rate': 0.05,
    'verbose': 0,
    'n_estimators': 200,
}
```

```python
train_data = lightgbm.Dataset(X_train, label=data_train['label'])
valid_data = lightgbm.Dataset(X_valid, label=data_valid['label'])
```

```python
model = lightgbm.train(parameters,
                       train_data,
                       valid_sets=valid_data,
                       num_boost_round=5000,
                       early_stopping_rounds=200)
```

```python
# clfs['gbm'] = model
# del clfs['gbm']
```

```python
df = pd.DataFrame()
for name, clf in clfs.items():
    df[name] = [a[1] for a in clf.predict_proba(X_valid)]
```

```python
tmp = model.predict(X_valid)
# rows = []
# for b in tmp:
#     rows.append([1-b, b])
df['bgm'] = [b for b in tmp]
```
